```python
with open('mapa.txt', 'r') as map:
    text = map.read()

cells = []

wall = []
for i in range(8):
    for j in range(6):
        new_wall = text[:4]
        wall.append(new_wall)
        text = text[5:]

alert = []
for i in range(3):
    pos_alert_x = text[0]
    pos_alert_y = text[2]
    pos_alert_state = text[4]
    text = text[6:]
    alert.append( (pos_alert_x, pos_alert_y, pos_alert_state) )

fire = []
for i in range(10):
    pos_fire_x = text[0]
    pos_fire_y = text[2]
    text = text[4:]
    fire.append( (pos_fire_x, pos_fire_y) )

door = []
for i in range(8):
    pos_doorA_x = text[0]
    pos_doorA_y = text[2]
    pos_doorB_x = text[4]
    pos_doorB_y = text[6]
    text = text[8:]
    door.append( ( (pos_doorA_x, pos_doorA_y), (pos_doorB_x, pos_doorB_y) ) )

exit = []
for i in range(4):
    pos_exit_x = text[0]
    pos_exit_y = text[2]
    text = text[4:]
    exit.append( (pos_exit_x, pos_exit_y) )
```
